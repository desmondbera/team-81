```python
def drop_last_row(df):
    return df.drop(df.tail(1).index)

def lower_case_col_names(df):
    return df.rename(columns=str.lower, inplace=True)

def rename_col_from(col_name, col_name_change, df):
    return df.rename(columns={col_name: col_name_change}, inplace=True)

#NOTE: we arrive to this number because 'R' is used for values between 1-5 (inclusive) but cannot be shown for privacy concerns. 
# As a result, we take the avg of 1+2+3+4+5 => 3
def change_r_occurence(df_race, df_gender, df_discipline, year):
    if(year == 2020):
        discipline_cols_with_r_val = ['principal', 'removal', 'superintendent',
           'sy1920_total_removals_suspensions']
        race_cols_with_r_val = ['american indian/alaskan native removals and/or suspensions resulting from incidents where nypd was contacted',
           'asian removals and/or suspensions resulting from incidents where nypd was contacted',
           'black removals and/or suspensions resulting from incidents where nypd was contacted',
           'hispanic removals and/or suspensions resulting from incidents where nypd was contacted',
           'multi-racial removals and/or suspensions resulting from incidents where nypd was contacted',
           'unknown removals and/or suspensions resulting from incidents where nypd was contacted',
           'white removals and/or suspensions resulting from incidents where nypd was contacted', 'sy1920_total_removals_suspensions removals and/or suspensions resulting from incidents where nypd was contacted'] 
        gender_cols_with_r_val = ['female removals and/or suspensions resulting from incidents where nypd was contacted',
           'male removals and/or suspensions resulting from incidents where nypd was contacted', 'sy1920_total_removals_suspensions removals and/or suspensions resulting from incidents where nypd was contacted']
        df_race[race_cols_with_r_val] = df_race[race_cols_with_r_val].replace(to_replace={'R': 3})
        df_gender[gender_cols_with_r_val] = df_gender[gender_cols_with_r_val].replace(to_replace={'R': 3})
        df_discipline[discipline_cols_with_r_val] = df_discipline[discipline_cols_with_r_val].replace(to_replace={'R': 3})
    elif(year == 2019):
        discipline_cols_with_r_val = ['removal', 'principal', 'superintendent', 'sy1819 total removals/suspensions']
        race_cols_with_r_val = ['american indian/alaskan native removals or suspensions resulting from incidents where nypd was contacted',
       'asian removals or suspensions resulting from incidents where nypd was contacted',
       'black removals or suspensions resulting from incidents where nypd was contacted',
       'hispanic removals or suspensions resulting from incidents where nypd was contacted',
       'white removals or suspensions resulting from incidents where nypd was contacted',
       'multi-racial removals or suspensions resulting from incidents where nypd was contacted',
       'unknown removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        gender_cols_with_r_val =['male removals or suspensions resulting from incidents where nypd was contacted',
       'female removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        df_race[race_cols_with_r_val] = df_race[race_cols_with_r_val].replace(to_replace={'R': 3})
        df_gender[gender_cols_with_r_val] = df_gender[gender_cols_with_r_val].replace(to_replace={'R': 3})
        df_discipline[discipline_cols_with_r_val] = df_discipline[discipline_cols_with_r_val].replace(to_replace={'R': 3})
    elif(year == 2018):
        print("year is 2018")
        discipline_cols_with_r_val = ['removal','principal', 'superintendent','sy1718 total removals/suspensions']
        race_cols_with_r_val = ['american indian/alaskan native removals or suspensions resulting from incidents where nypd was contacted',
       'asian removals or suspensions resulting from incidents where nypd was contacted',
       'black removals or suspensions resulting from incidents where nypd was contacted',
       'hispanic removals or suspensions resulting from incidents where nypd was contacted',
       'white removals or suspensions resulting from incidents where nypd was contacted',
       'multi-racial removals or suspensions resulting from incidents where nypd was contacted',
       'unknown removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        gender_cols_with_r_val = ['male removals or suspensions resulting from incidents where nypd was contacted',
       'female removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        
        df_race[race_cols_with_r_val] = df_race[race_cols_with_r_val].replace(to_replace={'R': 3})
        df_gender[gender_cols_with_r_val] = df_gender[gender_cols_with_r_val].replace(to_replace={'R': 3})
        df_discipline[discipline_cols_with_r_val] = df_discipline[discipline_cols_with_r_val].replace(to_replace={'R': 3})
    elif(year == 2017):
        print("year is 2017")
        discipline_cols_with_r_val = ['removal', 'principal', 'superintendent', 'expulsions', 'sy1718 total removals/suspensions']
        race_cols_with_r_val = ['american indian/alaskan native removals or suspensions resulting from incidents where nypd was contacted',
       'asian removals or suspensions resulting from incidents where nypd was contacted',
       'black removals or suspensions resulting from incidents where nypd was contacted',
       'hispanic removals or suspensions resulting from incidents where nypd was contacted',
       'white removals or suspensions resulting from incidents where nypd was contacted',
       'multi-racial removals or suspensions resulting from incidents where nypd was contacted',
       'unknown removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        gender_cols_with_r_val = ['male removals or suspensions resulting from incidents where nypd was contacted',
       'female removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        
        df_race[race_cols_with_r_val] = df_race[race_cols_with_r_val].replace(to_replace={'R': 3})
        df_gender[gender_cols_with_r_val] = df_gender[gender_cols_with_r_val].replace(to_replace={'R': 3})
        df_discipline[discipline_cols_with_r_val] = df_discipline[discipline_cols_with_r_val].replace(to_replace={'R': 3})
    elif(year == 2016):
        print("year is 2016")
        discipline_cols_with_r_val = ['removal', 'principal', 'superintendent', 'expulsions', 'sy1718 total removals/suspensions']
        race_cols_with_r_val = ['american indian/alaskan native removals or suspensions resulting from incidents where nypd was contacted',
       'asian removals or suspensions resulting from incidents where nypd was contacted',
       'black removals or suspensions resulting from incidents where nypd was contacted',
       'hispanic removals or suspensions resulting from incidents where nypd was contacted',
       'white removals or suspensions resulting from incidents where nypd was contacted',
       'multi-racial removals or suspensions resulting from incidents where nypd was contacted',
       'unknown removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        gender_cols_with_r_val = ['male removals or suspensions resulting from incidents where nypd was contacted',
       'female removals or suspensions resulting from incidents where nypd was contacted',
       'total removals or suspensions resulting from incidents where nypd was contacted']
        
        df_race[race_cols_with_r_val] = df_race[race_cols_with_r_val].replace(to_replace={'R': 3})
        df_gender[gender_cols_with_r_val] = df_gender[gender_cols_with_r_val].replace(to_replace={'R': 3})
        df_discipline[discipline_cols_with_r_val] = df_discipline[discipline_cols_with_r_val].replace(to_replace={'R': 3})
    else:
        print("This year does not exists in our change_r_occcurence function")
    return df_race, df_gender, df_discipline

def save_df_to_csv(df, file_name, year):
    df.to_csv(file_name+'_' + year +'.csv', encoding='utf-8')

```

```python
### 1. Datasets from 2020-2019 
#    -- 3 sheets in total
#    -- Result: 1 dataframe that has all of our relevant colum

xls_2020_discipline = pd.ExcelFile(url_2019_to_2020_student_discipline)
xls_2020_nypd_contacts = pd.ExcelFile(url_2019_to_2020_nypd_contact)

df_discipline_totals = pd.read_excel(xls_2020_discipline, index_col=0, sheet_name='Annual Report--R-P-S TOTALS', engine='openpyxl')
df_nypd_contact_race = pd.read_excel(xls_2020_nypd_contacts, index_col=0, sheet_name='NYPD-SUSP--RACE', engine='openpyxl')
df_nypd_contact_gender = pd.read_excel(xls_2020_nypd_contacts, index_col=0, sheet_name='NYPD-SUSP--GENDER', engine='openpyxl')

## 1 - DROP LAST ROW WITH NAN + TOTAL INFO
final_df_discipline_totals = drop_last_row(df_discipline_totals)
final_df_nypd_contact_race = drop_last_row(df_nypd_contact_race)
final_df_nypd_contact_gender = drop_last_row(df_nypd_contact_gender)

## 2 - LOWER CASE COL NAMES & REPLACE EMPTY SPACES WITH UNDERSCORES
# list(final_df_discipline_totals.columns)
lower_case_col_names(final_df_discipline_totals)
lower_case_col_names(final_df_nypd_contact_race)
lower_case_col_names(final_df_nypd_contact_gender)

## 3 - CHANGE COL NAMES TO BE MORE DESCRIPTIVE
rename_col_from('location_name', 'school_name', final_df_discipline_totals)
rename_col_from('location_category_description', 'school_type', final_df_discipline_totals)

rename_col_from('location_name', 'school_name', final_df_nypd_contact_race)
rename_col_from('location_category_description', 'school_type', final_df_nypd_contact_race)

rename_col_from('location_name', 'school_name', final_df_nypd_contact_gender)
rename_col_from('location_category_description', 'school_type', final_df_nypd_contact_gender)

## 4 - CHANGE 'R' occurences to int 3
df_tuple = change_r_occurence(final_df_nypd_contact_race, final_df_nypd_contact_gender, final_df_discipline_totals, 2020)
final_df_discipline_totals = df_tuple[2]
final_df_nypd_contact_race = df_tuple[0]
final_df_nypd_contact_gender = df_tuple[1]

## 5 - CHANGE index to unique nums
final_df_discipline_totals.reset_index(inplace=True)
final_df_nypd_contact_race.reset_index(inplace=True)
final_df_nypd_contact_gender.reset_index(inplace=True)

## 6 - MERGE 4 COLS FROM STUDENT_DISCIPLINE DF to NYPD_CONTACTS_RACE
final_df_nypd_contact_race.insert(11, 'principal', final_df_discipline_totals['principal'], True)
final_df_nypd_contact_race.insert(12, 'removal', final_df_discipline_totals['removal'], True)
final_df_nypd_contact_race.insert(13, 'superintendent', final_df_discipline_totals['superintendent'], True)
final_df_nypd_contact_race.insert(14, 'expulsions', final_df_discipline_totals['expulsions'], True)

## 7 - MERGE 2 NYPD_CONTACTS_GENDER COLS to NYPD_CONTACTS_RACE
final_df_nypd_contact_race.insert(15, 'female removals and/or suspensions resulting from incidents where nypd was contacted', final_df_nypd_contact_gender['female removals and/or suspensions resulting from incidents where nypd was contacted'], True)
final_df_nypd_contact_race.insert(16, 'male removals and/or suspensions resulting from incidents where nypd was contacted', final_df_nypd_contact_gender['male removals and/or suspensions resulting from incidents where nypd was contacted'], True)

save_df_to_csv(final_df_nypd_contact_race, 'clean_df', '2020-2019')


# print(final_df_discipline_totals.dtypes)
# print("====")
# print(final_df_nypd_contact_race.head())
# print("====")
# print(final_df_nypd_contact_gender.columns)
```
